```julia
using NPZ
using PyCall
using Glob
using PyPlot
using Random
using Eirene
using Statistics
using DelimitedFiles
```

```julia
tdaperyear = Any[]
for i in 1:18
    distance_pay = transpose(npzread("distancepay" * string(i) * ".npy"))
    results = eirene(distance_pay, model="pc", maxdim=2)
    push!(tdaperyear,results)
end
```

```julia
wassmatrix = Array{Float64}(undef, 18,18);
for i in 1:18
    for j in 1:18
        wassmatrix[i,j]= wasserstein_distance(barcode(tdaperyear[i], dim = 0),barcode(tdaperyear[j], dim = 0), p = 2, q = 2)
    end
end
```

```julia
max_coord = maximum(wassmatrix)
wassmatrix = wassmatrix/max_coord

figure(figsize=(10,8), dpi=300);
imshow(wassmatrix, cmap="gray", interpolation="none");
xticks(0:18-1, rotation=90);
yticks(0:18-1);
colorbar();
```
