```python
DATA_DIR = "../datasets_full"
```

```python
bt_df = pd.read_csv(os.path.join(DATA_DIR, 'Rio', 'brazil_tourism_jobs.csv'))
```

```python
bt_df['date'] = pd.to_datetime(bt_df[['year', 'month']].assign(Day=1))
```

```python
region_jobs = bt_df.pivot_table(index='date', columns='region', values='jobs', aggfunc=np.sum)
```

```python
region_jobs /= region_jobs.mean()
```
